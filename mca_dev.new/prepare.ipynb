```python
import pandas as pd
```

# Read

```python
enterobase = pd.read_csv("db/enterobase.tsv", sep='\t')
print(enterobase.columns)
enterobase = enterobase[['Assembly Barcode', 'SISTR1 Serovar', 'HC20', 'HC50', 'SISTR1 Serovar HC20', 'SISTR1 Serovar HC50', 'Source Niche', 'Collection Year']]
enterobase
```

```python
AMR_genes = pd.read_csv("db/AMRFinder_class.tsv", sep='\t')
AMR_genes
```

```python
amplicones = pd.read_csv("db/Abricate_barcode_aplicon.tsv", sep='\t')
amplicones
```

# Merge

```python
merged_enterobase_AMRFinder = pd.merge(AMR_genes, enterobase, on='Assembly Barcode', how='left')
merged_enterobase_AMRFinder.to_csv("out/genomas_con_AMR.csv", index=False)
merged_enterobase_AMRFinder
```

```python
merged_enterobase_amplicones = pd.merge(amplicones, enterobase, on='Assembly Barcode', how='left')
merged_enterobase_amplicones.to_csv("out/genomas_con_amplicones.csv", index=False)
merged_enterobase_amplicones
```

### Merge total

```python
merge = pd.merge(merged_enterobase_AMRFinder, amplicones, on='Assembly Barcode', how='left')
merge = merge[~merge.isnull().any(axis=1)] # elimino filas donde no hay amplicon o no hay AMR
#export = merge.drop(columns=['Assembly Barcode']).reset_index(drop=True)
merge.to_csv("out/genomas_con_amplicones_y_AMR.csv", index=False)
merge
```
