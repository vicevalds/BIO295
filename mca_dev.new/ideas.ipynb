```python
import pandas as pd
```

# Read

```python
AMRFinder = pd.read_csv("out/genomas_con_AMR.csv")
amplicones = pd.read_csv("out/genomas_con_amplicones.csv")
amplicones_y_AMR = pd.read_csv("out/genomas_con_amplicones_y_AMR.csv")
db = amplicones_y_AMR.drop(columns=['Assembly Barcode', 'HC50', 'SISTR1 Serovar HC50']).reset_index(drop=True)
db = db.rename(columns={'Source Niche': 'Niche', 'Collection Year': 'Year', 'Amplicon': 'Plasmid', 'SISTR1 Serovar': 'SISTR1'})
```

```python
db.to_csv("out-tests/all_table.csv", index=False)
db.tail(30)
#db["Class"].unique()
```

```python
"""
tabla solo por familias de AMR
"""
solo_por_familia = db.drop(columns=['Gene symbol', 'SISTR1 Serovar HC20']).reset_index(drop=True)
solo_por_familia.to_csv("out-tests/solo_por_familia.csv", index=False)
solo_por_familia
```

```python
"""
Solo el seropito agona y AMR por familias
"""
df = db.drop(columns=['Gene symbol', 'SISTR1 Serovar HC20']).reset_index(drop=True)
df = df[df['SISTR1'] == 'Agona']
df = df.drop(columns=['SISTR1']).reset_index(drop=True)
df.to_csv("out-tests/solo_agona_y_AMR_por_familia.csv", index=False)
df
```

```python
"""
Solo el seropito Typhimurium y AMR por familias
"""
df = db.drop(columns=['Gene symbol', 'SISTR1 Serovar HC20']).reset_index(drop=True)
df = df[df['SISTR1'] == 'Typhimurium']
df = df.drop(columns=['SISTR1']).reset_index(drop=True)
df.to_csv("out-tests/solo_Typhimurium_y_AMR_por_familia.csv", index=False)
df
```

```python
"""
Solo serotipos
"""
df = db.drop(columns=['Gene symbol', 'SISTR1 Serovar HC20', 'HC20']).reset_index(drop=True)
df.to_csv("out-tests/solo_serotipos_por_familia.csv", index=False)
```

```python
"""
Solo el seropito Infantis y AMR por familias
"""
df = db.drop(columns=['Gene symbol', 'SISTR1 Serovar HC20']).reset_index(drop=True)
df = df[df['SISTR1'] == 'Infantis']
df = df.drop(columns=['SISTR1', 'HC20']).reset_index(drop=True)
df.to_csv("out-tests/solo_Infantis_y_AMR_por_familia.csv", index=False)
df
```
