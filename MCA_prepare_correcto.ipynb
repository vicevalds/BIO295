```python
import pandas as pd
import os
```

```python
"""
Primero importar amr y filtrar por AMR de las familias de importancias, luego  

Tengo que tomar entero base, con sus barcode y quedar
"""
db = pd.read_csv(os.path.join("out", "amr.tsv"), sep='\t')
db = db.drop(columns=['HC50', 'SISTR1 Serovar HC20', 'SISTR1 Serovar HC50'])
db
```

```python
FAMILIAS = ['QUINOLONE', 'BETA-LACTAM', 'MACROLIDE', 'TRIMETHOPRIM']
SEROTIPOS = ['Typhimurium', 'Newport', 'Enteritidis', 'Infantis', 'I 1,4,[5],12:i:-', 'Agona']
```

### AMR

```python
amr = db[db['Class'].isin(FAMILIAS)]
print(amr.shape[0])
amr = amr[amr['SISTR1 Serovar'].isin(SEROTIPOS)]
print(amr.shape[0])
amr = amr.set_index('Assembly Barcode')
amr
```

### Abricate

```python
abricate = pd.read_csv("out/abricate_summary.tsv", sep='\t')
abricate = abricate.set_index("Assembly Barcode")
abricate.to_csv("XXX")
abricate
```

```python
nuevas_filas = []
for barcode, fila in abricate.iterrows():
    for gen, valor in fila.items():
        if valor == 1:
            nuevas_filas.append([barcode, gen])
nuevo_df = pd.DataFrame(nuevas_filas, columns=['Assembly Barcode', 'Amplicon'])
nuevo_df
```

# Merge n export

```python
df_merge = pd.merge(nuevo_df, amr, on='Assembly Barcode', how='right')
df_merge['Amplicon'] = df_merge['Amplicon'].fillna('No_plasmid')

df_merge = df_merge.rename(columns={'Resistance gene': 'AMR', 'SISTR1 Serovar': 'Serovar', 'Source Niche': 'Niche', 'Collection Year': 'Year'})

df_merge = df_merge.drop(columns=['Assembly Barcode', 'AMR'])
df_merge.to_csv("out/MCA_table_v1.csv", index=False)
df_merge
```
